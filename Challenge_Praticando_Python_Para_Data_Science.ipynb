### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
loja2.head()
loja3.head()
loja4.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
print('Faturamento da Loja 1:', faturamento_loja1)
faturamento_loja2 = loja2['Preço'].sum()
print('Faturamento da Loja 2:', faturamento_loja2)
faturamento_loja3 = loja3['Preço'].sum()
print('Faturamento da Loja 3:', faturamento_loja3)
faturamento_loja4 = loja4['Preço'].sum()
print('Faturamento da Loja 4:', faturamento_loja4)
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
```

```python
import matplotlib.pyplot as plt
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
plt.figure(figsize=(10, 5))
plt.bar(lojas, faturamento)
plt.title('Faturamento')
plt.ylabel('Preço')
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
vendas_por_categoria_loja1 = loja.groupby('Categoria do Produto') \
                         .size().reset_index(name='Quantidade de Vendas')
print(vendas_por_categoria_loja1)
quantidade_vendas_loja1 = vendas_por_categoria_loja1['Quantidade de Vendas'].sum()
print("Total de vendas da loja 1:", quantidade_vendas_loja1)
```

```python
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto') \
                         .size().reset_index(name='Quantidade de Vendas')
print(vendas_por_categoria_loja2)
quantidade_vendas_loja2 = vendas_por_categoria_loja2['Quantidade de Vendas'].sum()
print("Total de vendas da loja 2:", quantidade_vendas_loja2)
```

```python
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto') \
                         .size().reset_index(name='Quantidade de Vendas')
print(vendas_por_categoria_loja3)
quantidade_vendas_loja3 = vendas_por_categoria_loja3['Quantidade de Vendas'].sum()
print("Total de vendas da loja 3:", quantidade_vendas_loja3)
```

```python
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto') \
                         .size().reset_index(name='Quantidade de Vendas')
print(vendas_por_categoria_loja4)
quantidade_vendas_loja4 = vendas_por_categoria_loja4['Quantidade de Vendas'].sum()
print("Total de vendas da loja 4:", quantidade_vendas_loja4)
```

```python
vendas_ordenadas_loja1 = vendas_por_categoria_loja1.sort_values('Quantidade de Vendas', ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(vendas_ordenadas_loja1['Quantidade de Vendas'], labels=vendas_ordenadas_loja1['Categoria do Produto'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vendas por Categoria de Produto da Loja 1')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
vendas_ordenadas_loja2 = vendas_por_categoria_loja2.sort_values('Quantidade de Vendas', ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(vendas_ordenadas_loja2['Quantidade de Vendas'], labels=vendas_ordenadas_loja2['Categoria do Produto'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vendas por Categoria de Produto da Loja 2')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
vendas_ordenadas_loja3 = vendas_por_categoria_loja3.sort_values('Quantidade de Vendas', ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(vendas_ordenadas_loja3['Quantidade de Vendas'], labels=vendas_ordenadas_loja3['Categoria do Produto'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vendas por Categoria de Produto da Loja 3')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
vendas_ordenadas_loja4 = vendas_por_categoria_loja4.sort_values('Quantidade de Vendas', ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(vendas_ordenadas_loja4['Quantidade de Vendas'], labels=vendas_ordenadas_loja4['Categoria do Produto'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vendas por Categoria de Produto da Loja 4')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_total = []
loja['Avaliação da compra'] = pd.to_numeric(loja['Avaliação da compra'], errors='coerce')
media_total_loja1 = loja['Avaliação da compra'].mean()
print(f"Média total das avaliações da loja 1: {media_total_loja1:.2f}")
loja2['Avaliação da compra'] = pd.to_numeric(loja2['Avaliação da compra'], errors='coerce')
media_total_loja2 = loja2['Avaliação da compra'].mean()
print(f"Média total das avaliações da loja 2: {media_total_loja2:.2f}")
loja3['Avaliação da compra'] = pd.to_numeric(loja3['Avaliação da compra'], errors='coerce')
media_total_loja3 = loja3['Avaliação da compra'].mean()
print(f"Média total das avaliações da loja 3: {media_total_loja3:.2f}")
loja4['Avaliação da compra'] = pd.to_numeric(loja4['Avaliação da compra'], errors='coerce')
media_total_loja4 = loja4['Avaliação da compra'].mean()
print(f"Média total das avaliações da loja 4: {media_total_loja4:.2f}")
media_total = [media_total_loja1, media_total_loja2, media_total_loja3, media_total_loja4]
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
plt.figure(figsize=(10, 5))
plt.barh(lojas, media_total)
plt.title('Média das Avaliações')
plt.tight_layout()
plt.show()
```

```python
vendas_por_produto_loja3 = loja3['Produto'].value_counts().reset_index()
vendas_por_produto_loja3.columns = ['Produto', 'Quantidade de Vendas']
plt.figure(figsize=(8,10))
plt.barh(vendas_por_produto_loja3['Produto'], vendas_por_produto_loja3['Quantidade de Vendas'], color='skyblue')
plt.title('Quantidade de Vendas por Produto da Loja 3')
plt.xlabel('Quantidade de Vendas')
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_total = []
loja['Frete'] = pd.to_numeric(loja['Frete'], errors='coerce')
frete_medio_loja1 = loja['Frete'].mean()
print(f"Custo médio de frete da loja 1: R$ {frete_medio_loja1:.2f}")
loja2['Frete'] = pd.to_numeric(loja2['Frete'], errors='coerce')
frete_medio_loja2 = loja2['Frete'].mean()
print(f"Custo médio de frete da loja 2: R$ {frete_medio_loja2:.2f}")
loja3['Frete'] = pd.to_numeric(loja3['Frete'], errors='coerce')
frete_medio_loja3 = loja3['Frete'].mean()
print(f"Custo médio de frete da loja 3: R$ {frete_medio_loja3:.2f}")
loja4['Frete'] = pd.to_numeric(loja4['Frete'], errors='coerce')
frete_medio_loja4 = loja4['Frete'].mean()
print(f"Custo médio de frete da loja 4: R$ {frete_medio_loja4:.2f}")
frete_total = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
plt.figure(figsize=(10, 5))
plt.barh(lojas, frete_total)
plt.title('Média dos Fretes')
plt.tight_layout()
plt.show()
```
